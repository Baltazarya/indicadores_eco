```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

# Estilo dos gráficos
sns.set_theme(style="whitegrid", palette="muted", font_scale=1.2)

# Carregar variáveis de ambiente (.env)
load_dotenv()

# Conexão com PostgreSQL
usuario = os.getenv('DB_USER')
senha = os.getenv('DB_PASSWORD')
host = os.getenv('DB_HOST')
porta = os.getenv('DB_PORT')
banco = os.getenv('DB_NAME')

url = f'postgresql+psycopg2://{usuario}:{senha}@{host}:{porta}/{banco}'
engine = create_engine(url)
```

```python
# Carregar dados da tabela no banco
query = "SELECT * FROM indicadores_desemprego"
df = pd.read_sql(query, engine)
```

```python
# Plotando os dados
variaveis = df.columns.drop('data')

plt.figure(figsize=(15, 25))

for i, var in enumerate(variaveis, 1):
    plt.subplot(len(variaveis), 1, i)
    plt.plot(df['data'], df[var], label=var)
    plt.title(f'{var} ao longo do tempo')
    plt.xlabel('Data')
    plt.ylabel(var)
    plt.grid(True)
    plt.tight_layout()

plt.show()
```

# 📊 Análise Temporal dos Indicadores Econômicos

A seguir, apresentamos uma análise descritiva dos indicadores econômicos brasileiros ao longo do tempo, com base nos gráficos fornecidos.

---

## 📈 1. Taxa de Desemprego (`taxa_desemprego`)

- **Tendência**: Alta entre 2015 e 2020, com pico por volta de 2020–2021. Queda constante a partir de 2022.
- **Possível Interpretação**: A alta reflete o impacto da crise econômica e da pandemia. A queda pode estar relacionada à retomada econômica e políticas de estímulo.

---

## 👥 2. População Desocupada (`pop_desocupada`)

- **Tendência**: Queda de 2015 a 2017, estabilidade até 2020, seguida de aumento e novo recuo recente.
- **Relação com o Desemprego**: Altamente correlacionada com a taxa de desemprego, como esperado.

---

## 👷 3. População Ocupada (`pop_ocupada`)

- **Tendência**: Queda de 2015 até 2020, recuperação contínua até 2025.
- **Interpretação**: Mostra recuperação do mercado de trabalho pós-pandemia, em sintonia com a queda do desemprego.

---

## 💵 4. Rendimento Médio (`rendimento_medio`)

- **Tendência**: Estável até 2020, crescimento acentuado a partir de 2021.
- **Possível Interpretação**: Pode indicar reajustes salariais, valorização do salário mínimo, aumento na produtividade ou efeitos inflacionários.

---

## 🏭 5. Produto Interno Bruto (`pib`)

- **Tendência**: Crescimento consistente ao longo do período, sem grandes quedas aparentes.
- **Observação**: Indica crescimento econômico de longo prazo, ainda que moderado.

---

## 💰 6. Taxa SELIC (`selic`)

- **Tendência**: Queda de 2016 a 2020, mínima histórica por volta de 2020, seguida de forte alta até 2023.
- **Possível Interpretação**: Política monetária expansionista durante a pandemia, seguida de aumento para conter a inflação.

---

## 📈 7. IPCA (Inflação) (`ipca`)

- **Tendência**: Bastante volátil, com picos nos anos de 2021–2022.
- **Relação com SELIC**: O aumento da SELIC pode ser resposta ao aumento do IPCA no período.

---

## 🤝 8. Confiança do Consumidor (`confianca_consumidor`)

- **Tendência**: Alta entre 2015 e 2018, queda até 2020, alta instável depois da pandemia.
- **Interpretação**: Sensível a crises e incertezas políticas/econômicas. A oscilação pós-pandemia indica recuperação ainda incerta.

---

## 🏗️ 9. Produção Industrial (`producao_industrial`)

- **Tendência**: Muito volátil, com queda expressiva em 2020 e recuperação parcial posterior.
- **Possível Interpretação**: Impactada fortemente por choques externos (pandemia) e variações na demanda interna/externa.

---

## Conclusão

O ano de 2020 marca uma inflexão significativa nos indicadores, refletindo os impactos da pandemia. A partir de 2021, observa-se uma tendência de recuperação em diversos indicadores, como PIB, ocupação e confiança do consumidor. A elevação da SELIC surge como resposta à inflação, enquanto a produção industrial e o rendimento médio seguem trajetórias mais instáveis. Em conjunto, os dados mostram que podemos ter em maõs um cenário de transição e reaquecimento da economia brasileira.
